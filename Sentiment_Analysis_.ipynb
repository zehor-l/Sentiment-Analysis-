```python
# Load the sentiment analysis model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)
model = transformers.AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Define a function for sentiment analysis
def predict_sentiment(text):
    inputs = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        return_tensors="pt"
    )

    # Feed the input to the model and get the predicted class
    outputs = model(**inputs)
    logits = outputs.logits
    probs = torch.softmax(logits, dim=1)
    predicted_class = torch.argmax(probs, dim=1)
    return predicted_class.item()
```

```python
# Test the function on some example texts
texts = [
    "I love this product!",
    "My exceperience is terrible.",
    "The app was bad.",
    "I am disapointed by this this service."
]

for text in texts:
    sentiment = predict_sentiment(text)
    if sentiment == 0:
        print(f"{text} - Negative")
    elif sentiment == 1:
        print(f"{text} - Positive")
    else:
        print(f"{text} - Neutral")
```
